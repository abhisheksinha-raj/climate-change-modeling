# üåç My Climate Change Adventure!

Hi! I'm a climate scientist studying how our planet is getting warmer. In this project, I‚Äôll:
- Look at temperature data from 2000 to 2020.
- Make cool graphs to see the warming trend.
- Predict what the Earth‚Äôs temperature might be in 2025 and 2030.


Let‚Äôs dive into the data! üöÄ

[Learn more about climate change here!](https://www.nasa.gov/kidsclub/climate_change.html)

```python
import pandas as pd
import numpy as np

# Make fake temperature data for 2000 to 2020
data = {
    'Year': list(range(2000, 2021)),
    'Temperature': [14.5, 14.6, 14.7, 14.8, 14.9, 15.0, 15.1, 15.2, 15.3, 15.4,
                   15.5, 15.6, 15.7, 15.8, 15.9, 16.0, 16.1, 16.2, 16.3, 16.4, 16.5]
}

# Put data in a table
df = pd.DataFrame(data)

# Save it to a file
df.to_csv('temperature_data.csv', index=False)

# Show the first few rows
print(df.head())
```

## üìä My Temperature Data

I made a table with years (2000‚Äì2020) and fake temperatures to see how the Earth is warming. Each number shows the average global temperature in Celsius. For example, 14.5¬∞C in 2000 means it was a bit chilly, but 16.5¬∞C in 2020 shows it‚Äôs getting hotter! üî•

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Make a scatter plot
plt.figure(figsize=(10, 5))
sns.scatterplot(x='Year', y='Temperature', data=df, color='blue', s=100)
plt.title('Global Temperature Over Time')
plt.xlabel('Year')
plt.ylabel('Temperature (¬∞C)')
plt.grid(True)
plt.savefig('temperature_trend.png')
plt.show()
```

## üìà Cool Graphs of Warming Temperatures

Look at my graph! The blue dots show how the Earth‚Äôs temperature went up from 2000 to 2020. The line in the chart above connects the dots to show the warming trend. It‚Äôs like the Earth is getting a fever! üòì

![Temperature Trend](temperature_trend.png)

**Why It Matters**: This graph shows the Earth is getting warmer every year. We need to help cool it down by planting trees and using less energy!

Want to see a fun video about climate change? [Watch this!](https://www.youtube.com/watch?v=G4H1N_yXBiA)

```python
from sklearn.linear_model import LinearRegression

# Prepare data for prediction
X = df[['Year']].values  # Years
y = df['Temperature'].values  # Temperatures

# Create and train the model
model = LinearRegression()
model.fit(X, y)

# Predict for 2025 and 2030
future_years = np.array([[2025], [2030]])
predictions = model.predict(future_years)

# Show predictions
print("Predicted temperatures:")
for year, temp in zip([2025, 2030], predictions):
    print(f"Year {year}: {temp:.1f}¬∞C")

# Draw a graph with predictions
plt.figure(figsize=(10, 5))
sns.scatterplot(x='Year', y='Temperature', data=df, color='blue', s=100, label='Past Data')
plt.plot(df['Year'], model.predict(X), color='red', label='Trend Line')
plt.scatter(future_years, predictions, color='green', marker='*', s=200, label='Predictions')
plt.title('Temperature Trend with Predictions')
plt.xlabel('Year')
plt.ylabel('Temperature (¬∞C)')
plt.legend()
plt.grid(True)
plt.savefig('temperature_predictions.png')
plt.show()
```

## üîÆ Predicting the Future!

I used a magic tool called *Linear Regression* to guess what the Earth‚Äôs temperature might be in 2025 and 2030. The green stars in the graph show my predictions, and the red line shows the warming trend. It looks like the Earth will keep getting hotter! üå°Ô∏è
